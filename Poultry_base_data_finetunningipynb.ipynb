```python
import os, fitz, docx2txt, sys
from pptx import Presentation
from tqdm import tqdm
from pathlib import Path

OUT_PATH = "/content/drive/MyDrive/poultry_text"
os.makedirs(OUT_PATH, exist_ok=True)

def extract_file(fp):
    try:
        if fp.lower().endswith(".pdf"):
            doc = fitz.open(fp)
            return "".join([p.get_text() for p in doc])
        elif fp.lower().endswith(".pptx"):
            prs = Presentation(fp)
            return "\n".join([s.text for slide in prs.slides for s in slide.shapes if hasattr(s,"text")])
        elif fp.lower().endswith(".docx"):
            return docx2txt.process(fp)
    except:
        return ""

# collect all files first
all_files = []
for r,_,fs in os.walk(BASE_PATH):
    for f in fs:
        if f.lower().endswith((".pdf",".pptx",".docx")):
            all_files.append(os.path.join(r,f))

print("Total files:", len(all_files))

bar = tqdm(total=len(all_files), ncols=90)

for i,fp in enumerate(all_files,1):
    bar.set_description(f"[{i}/{len(all_files)}] {Path(fp).name[:45]}")
    txt = extract_file(fp)
    if txt.strip():
        open(os.path.join(OUT_PATH, Path(fp).name + ".txt"),"w",encoding="utf-8").write(txt)
    bar.update(1)

bar.close()
print("Completed")
```

```python
#data cleaning
import re, os
from tqdm import tqdm
from pathlib import Path

RAW_PATH = "/content/drive/MyDrive/poultry_text"
CLEAN_PATH = "/content/drive/MyDrive/poultry_clean"
os.makedirs(CLEAN_PATH, exist_ok=True)

def clean_text(t):
    t = re.sub(r'\n+', '\n', t)
    t = re.sub(r'Page\s*\d+', '', t, flags=re.I)
    t = re.sub(r'\d+\s*/\s*\d+', '', t)
    t = re.sub(r'_{3,}', '', t)
    t = re.sub(r'\s{2,}', ' ', t)
    return t.strip()

files = list(Path(RAW_PATH).glob("*.txt"))
print("Cleaning files:", len(files))

for f in tqdm(files):
    raw = open(f,encoding="utf-8",errors="ignore").read()
    clean = clean_text(raw)
    if len(clean.split()) > 100:
        open(os.path.join(CLEAN_PATH, f.name),"w",encoding="utf-8").write(clean)
        (clean)

print("Clean text ready")
```

```python

#chunking
import os, json
from tqdm import tqdm
from pathlib import Path

TEXT_PATH = "/content/drive/MyDrive/poultry_text"
CHUNK_PATH = "/content/drive/MyDrive/poultry_chunks"
os.makedirs(CHUNK_PATH, exist_ok=True)

MAX_WORDS = 500
OVERLAP = 80

def chunk_text(text):
    words = text.split()
    chunks = []
    start = 0
    while start < len(words):
        end = start + MAX_WORDS
        chunk = " ".join(words[start:end])
        chunks.append(chunk)
        start += MAX_WORDS - OVERLAP
    return chunks

all_txt = list(Path(TEXT_PATH).glob("*.txt"))
print("Total text files:", len(all_txt))

cid = 0
for f in tqdm(all_txt):
    text = open(f, encoding="utf-8", errors="ignore").read()
    for chunk in chunk_text(text):
        out = {
            "id": cid,
            "source": f.name,
            "text": chunk
        }
        open(f"{CHUNK_PATH}/{cid}.json","w",encoding="utf-8").write(json.dumps(out))
        cid += 1

print("Total chunks created:", cid)
```

```python
#embedding 2
import os, json
import faiss
from tqdm import tqdm
from sentence_transformers import SentenceTransformer
from pathlib import Path
import numpy as np

CHUNK_PATH = "/content/drive/MyDrive/poultry_chunks"
FAISS_PATH = "/content/drive/MyDrive/poultry_faiss"
os.makedirs(FAISS_PATH, exist_ok=True)


model = SentenceTransformer("all-MiniLM-L6-v2")

chunks = []
meta = []

files = sorted(Path(CHUNK_PATH).glob("*.json"))
print("Total chunks:", len(files))

for f in tqdm(files):
    obj = json.load(open(f))
    chunks.append(obj["text"])
    meta.append({
        "id": obj["id"],
        "source": obj["source"]
    })

# bacth embedding
embeddings = model.encode(
    chunks,
    batch_size=32,
    show_progress_bar=True,
    normalize_embeddings=True
)

dim = embeddings.shape[1]
index = faiss.IndexFlatIP(dim)
index.add(np.array(embeddings))

# save brain
faiss.write_index(index, os.path.join(FAISS_PATH, "poultry.index"))
json.dump(meta, open(os.path.join(FAISS_PATH, "meta.json"), "w"))

print("FAISS Done")
```

```python
import os, json, faiss
import numpy as np
from tqdm import tqdm
from sentence_transformers import SentenceTransformer
from pathlib import Path

# where your old brain lives
OLD_BRAIN = "/content/drive/MyDrive/poultry_faiss"
# where new chunks to be added live
NEW_CHUNKS = "/content/drive/MyDrive/poultry_new_chunks"

print("Loading old brain...")

# load existing FAISS brain
index = faiss.read_index(os.path.join(OLD_BRAIN, "poultry.index"))
meta = json.load(open(os.path.join(OLD_BRAIN, "meta.json")))

# load embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")

new_embeddings = []
new_meta = []

files = list(Path(NEW_CHUNKS).glob("*.json"))
print("New chunks found:", len(files))

for f in tqdm(files):
    obj = json.load(open(f))
    new_embeddings.append(obj["text"])
    new_meta.append({
        "id": obj["id"],
        "source": obj["source"]
    })

# convert text → vectors
vecs = model.encode(new_embeddings, batch_size=32, normalize_embeddings=True)

print("Adding new knowledge to brain...")

# append new vectors into FAISS brain
index.add(np.array(vecs))

# append metadata
meta.extend(new_meta)

# save updated brain back to Drive
faiss.write_index(index, os.path.join(OLD_BRAIN, "poultry.index"))
json.dump(meta, open(os.path.join(OLD_BRAIN, "meta.json"), "w"))

print("🧠 Brain updated. New total memories:", index.ntotal)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, TrainingArguments
from peft import LoraConfig, get_peft_model
from trl import SFTTrainer


model_id = "microsoft/phi-3-mini-4k-instruct"
tokenizer = AutoTokenizer.from_pretrained(model_id)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    load_in_4bit=True,
    device_map="auto"
)

lora = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["q_proj","v_proj"],
    lora_dropout=0.05,
    task_type="CAUSAL_LM"
)

model = get_peft_model(model, lora)

trainer = SFTTrainer(
    model=model,
    train_dataset=data["train"],
    dataset_text_field="text",
    max_seq_length=512,
    args=TrainingArguments(
        per_device_train_batch_size=2,
        gradient_accumulation_steps=4,
        num_train_epochs=2,
        output_dir="/content/drive/MyDrive/poultry_cpt_model",
        save_strategy="epoch",
        logging_steps=50,
        learning_rate=2e-4,
        fp16=True
    )
)

trainer.train()
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, TrainingArguments
from peft import LoraConfig, get_peft_model
from trl import SFTTrainer

model_id = "microsoft/phi-3-mini-4k-instruct"
tokenizer = AutoTokenizer.from_pretrained(model_id)
tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    load_in_4bit=True,
    device_map="auto",
    trust_remote_code=True
)

lora = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["qkv_proj","o_proj"],
    lora_dropout=0.05,
    task_type="CAUSAL_LM"
)
model = get_peft_model(model, lora)

trainer = SFTTrainer(
    model=model,
    train_dataset=data["train"],
    dataset_text_field="text",
    max_seq_length=512,
    args=TrainingArguments(
        per_device_train_batch_size=2,
        gradient_accumulation_steps=4,
        num_train_epochs=2,
        output_dir="/content/drive/MyDrive/poultry_cpt_model",
        save_strategy="epoch",
        logging_steps=50,
        learning_rate=2e-4,
        fp16=True
    )
)

trainer.train()
```
